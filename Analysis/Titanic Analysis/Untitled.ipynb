```python
import requests
import pandas as pd

# Get the current season's ID
url = "https://fantasy.premierleague.com/api/bootstrap-static/"
response = requests.get(url)
data = response.json()
current_gameweek = max([int(event['id']) for event in data['events'] if event['finished'] is True])

# Get the standings for the current gameweek
url = f"https://fantasy.premierleague.com/api/event/{current_gameweek}/live/"
response = requests.get(url)
data = response.json()
standings = data['elements']

# Create a list of dictionaries containing the relevant data for each team
teams = []
for team in standings:
    team_data = {}
#     team_data['name'] = team['team_name']
    team_data['played'] = team['stats']['minutes'] / 90
    team_data['won'] = team['stats']['wins']
    team_data['drawn'] = team['stats']['draws']
    team_data['lost'] = team['stats']['losses']
    team_data['points'] = team['stats']['total_points']
    team_data['goals_for'] = team['stats']['goals_scored']
    team_data['goals_against'] = team['stats']['goals_conceded']
    teams.append(team_data)

# Create a Pandas DataFrame from the list of team dictionaries
table = pd.DataFrame(teams)

# Sort the DataFrame by points, then by goal difference, then by goals scored
table = table.sort_values(['points', 'goals_for', 'goals_against'], ascending=[False, False, True])

# Add a position column to the DataFrame
table['position'] = range(1, len(table) + 1)

# Print the table
print(table)
```
