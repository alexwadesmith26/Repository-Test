```python
data_names = ["ghg_data", "pop_data", "gdp_data"]
datasets = ["ghg_data.csv", "pop_data.csv", "gdp_data.csv"]
d = {name: pd.read_csv(data) for name, data in zip(data_names, datasets)}

# [x + y for x,y in zip(l1,l2)]
```

```python
variable_names = ["GHG Emissions", "Population", "GDP"]

G7_countries = ['Canada', 'France', 'Germany', 'Italy', 'Japan', 'United Kingdom', 'United States','New Zealand']

G20_countries = ["Argentina", "Australia", "Brazil", "Canada", "China"
                 , "France", "Germany", "India", "Indonesia", "Italy"
                 , "Japan", "Mexico", "Russian Federation", "Saudi Arabia", "South Africa"
                 , "Korea, Rep.", "Turkiye", "United Kingdom", "United States"]

for i, k in enumerate(data_names):
    d[k] = d[k].drop(['Series Name', 'Series Code'], axis=1)
    d[k] = pd.melt(d[k], id_vars = 'Country Name', value_vars = d[k].columns[2:])
    d[k] = d[k].dropna()
    d[k] = d[k].loc[d[k]["value"] != ".." ]
    d[k]["value"] = pd.to_numeric(d[k]["value"])
    d[k]['Year'] = d[k]["variable"].apply(lambda x: pd.Series(str(x).split(" ")[0]))
    d[k] = d[k].drop('variable', axis=1)
    d[k] = d[k].rename(columns={"value":variable_names[i]})
    d[k] = d[k].reset_index(drop=True)
    d[k][variable_names[i] + " (million)"] = d[k][variable_names[i]] / 1000000
    d[k]['is_G7'] = np.where(d[k]['Country Name'].isin(G7_countries), True, False)
    d[k]['is_G20'] = np.where(d[k]['Country Name'].isin(G20_countries), True, False)
    d[k]['Year'] = pd.to_numeric(d[k]['Year'])
```

```python
df_comp = pd.merge(d['ghg_data'], d['pop_data'], how='inner'
                   , left_on=['Country Name','Year', 'is_G7', 'is_G20'], right_on=['Country Name','Year', 'is_G7', 'is_G20'])

df_comp = pd.merge(df_comp, d['gdp_data'], how='inner'
                   , left_on=['Country Name','Year', 'is_G7', 'is_G20'], right_on=['Country Name','Year', 'is_G7', 'is_G20'])
```

```python
df_comp['GHG per cap'] = df_comp['GHG Emissions'] / df_comp['Population']
df_comp['GHG per million GDP'] = df_comp['GHG Emissions'] / df_comp['GDP (million)']
```

```python
GHG = np.array(d['ghg_data'][d['ghg_data']['Year'] == 2019]["GHG Emissions (million)"])

# Different sades of grey used in the plot
GREY88 = "#e0e0e0"
GREY85 = "#d9d9d9"
GREY82 = "#d1d1d1"
GREY79 = "#c9c9c9"
GREY97 = "#f7f7f7"
GREY60 = "#999999"

# Values for the x axis
ANGLES = np.linspace(0, 2 * np.pi, len(GHG), endpoint=False)

# Heights of the lines and y-position of the dot are given by the times.
HEIGHTS = np.array(GHG) + 0.000000001

PLUS = 2
```

```python
GHG_MAX = np.max(GHG)
GHG_MIN = np.min(GHG)

# 'low' and 'high' refer to the final dot size.
def scale_to_interval(x, low=1, high=60):
    return ((x - GHG_MIN) / (GHG_MAX - GHG_MIN)) * (high - low) + low
```

```python
# Initialize layout in polar coordinates
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw={"projection": "polar"})

# Set background color to white, both axis and figure.
fig.patch.set_facecolor("white")
ax.set_facecolor("white")

# Use logarithmic scale for the radial axis
# ax.set_rscale('log', basex= 2)

# Angular axis starts at 90 degrees, not at 0
ax.set_theta_offset(np.pi / 2)

# Reverse the direction to go counter-clockwise.
ax.set_theta_direction(-1)

# Add lines
ax.vlines(ANGLES, 0 + PLUS, HEIGHTS + PLUS, lw=0.9)

# Add dots
ax.scatter(ANGLES, HEIGHTS + PLUS, s=scale_to_interval(HEIGHTS));
```
